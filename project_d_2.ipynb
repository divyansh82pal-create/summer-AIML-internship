```python
data["Age"].fillna(data["Age"].median(),inplace=True)
data["Income"].fillna(data["Income"].median())
data["Spending_Score"].fillna(data["Spending_Score"].median())
data["City"].fillna(data["City"].mode()[0])
print(data)
```

```python
print(data["Gender"].mode())
```

```python
print(data["City"].mode())
```

```python
data["Gender"].fillna(data["Gender"].mode()[0],inplace=True)
data["City"].fillna(data["City"].mode()[0],inplace=True)
print(data)
```

```python
data.drop_duplicates(inplace=True)
data.head(7)
```

```python
data["Purchased"]=data["Purchased"].astype("category")
print(data)
```

```python
Q1=data['Spending_Score'].quantile(0.25)
Q3=data['Spending_Score'].quantile(0.75)
IQR=Q3-Q1
print(Q1,Q3,IQR)
```
